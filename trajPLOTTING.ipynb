```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV file
df = pd.read_csv('Trajectories/dq/trajectoryHop13dq.csv', header=None)

# Assign column names
column_names = [f'q{i}' for i in range(1, 13)] + ['duration'] + [f'dq{i}' for i in range(1, 13)]
df.columns = column_names

# Calculate cumulative time
df['cumulative_time'] = df['duration'].cumsum()

# Define mapping for descriptive names
q_names = {
    'q2': 'Front Thigh',
    'q3': 'Front Calf',
    'q8': 'Rear Thigh',
    'q9': 'Rear Calf'
}

dq_names = {
    'dq2': 'Front Thigh Velocity',
    'dq3': 'Front Calf Velocity',
    'dq8': 'Rear Thigh Velocity',
    'dq9': 'Rear Calf Velocity'
}

# Create the plot
plt.figure(figsize=(15, 10))

# Plot inverted q values
plt.subplot(2, 1, 1)
for q, name in q_names.items():
    plt.plot(df['cumulative_time'], -df[q], label=f'-{name}')
plt.title('Inverted Joint Angles over Time')
plt.xlabel('Time (s)')
plt.ylabel('Inverted Angle (rad)')
plt.legend()

# Plot inverted dq values
plt.subplot(2, 1, 2)
for dq, name in dq_names.items():
    plt.plot(df['cumulative_time'], -df[dq], label=f'-{name}')
plt.title('Inverted Joint Velocities over Time')
plt.xlabel('Time (s)')
plt.ylabel('Inverted Angular Velocity (rad/s)')
plt.legend()

plt.tight_layout()
plt.show()

# Print statistics (using original, non-inverted values)
print("\nStatistics for Joint Angles - Original Values:")
for q, name in q_names.items():
    print(f"{name} - Min: {df[q].min():.4f}, Max: {df[q].max():.4f}, Mean: {df[q].mean():.4f}")

print("\nStatistics for Joint Velocities - Original Values:")
for dq, name in dq_names.items():
    print(f"{name} - Min: {df[dq].min():.4f}, Max: {df[dq].max():.4f}, Mean: {df[dq].mean():.4f}")

print(f"\nTotal duration: {df['cumulative_time'].max():.4f} seconds")

# Calculate and print the range of motion for each joint (using original values)
print("\nRange of Motion for each joint:")
for q, name in q_names.items():
    range_of_motion = df[q].max() - df[q].min()
    print(f"{name} - Range of Motion: {range_of_motion:.4f} radians")

# Calculate and print the maximum velocity for each joint (using original values)
print("\nMaximum Velocity for each joint:")
for dq, name in dq_names.items():
    max_velocity = df[dq].abs().max()
    print(f"{name} - Max Velocity: {max_velocity:.4f} rad/s")

# Print information about the effect of inversion
print("\nNote on Inversion:")
print("The plots show the negative values of the original data.")
print("This inversion changes the direction of motion in the visualization:")
print("- Positive angles become negative, and vice versa.")
print("- Clockwise rotations may appear counterclockwise, and vice versa.")
print("- The shape of the curves remains the same, but mirrored vertically.")
```
